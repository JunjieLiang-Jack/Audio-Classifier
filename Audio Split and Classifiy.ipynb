```python
#check audio file subtype, "IMA ADPCM" will not work with pybub, need to load it into sf and conver it into PCM_16.
f = sf.SoundFile(os.path.join(ADPCM_folder,filename))
print(f.format, f.subtype)

#load it into sf and rewrite to PCM_16
data, samplerate = sf.read(os.path.join(ADPCM_folder,filename))
sf.write(os.path.join(PCM_folder,filename), data, samplerate)
```

```python
#AudioSegment.converter = which("ffmpeg")
myaudio = AudioSegment.from_file(os.path.join(PCM_folder,filename), format="wav") 
Split_length_ms = 4000 # pydub calculates in millisec 
Splits = make_chunks(myaudio,Split_length_ms) #Make Splits of four sec 
for i, Split in enumerate(Splits): 
    Split_name = "{0}.wav".format(i) 
    print ("exporting", Split_name) 
    Split.export(os.path.join(output_file_path,Split_name), format="wav") 

print("Total number of Split = ",len(Splits))
```

```python
!git clone https://github.com/GorillaBus/urban-audio-classifier.git
import sys
sys.path.insert(0,'/content/urban-audio-classifier/include')
import helpers
from IPython.display import clear_output, display


#testing data need to extract the Mel_spec inorder to be predicted by the modle. 
features = []
labels = []
counter = 0
n_mels = 40
frames_max = 0
total_test_samples = len(os.listdir(output_file_path))

for filename in os.listdir(output_file_path):
  if filename.endswith(".wav"):
    file_path = os.path.join(output_file_path,filename)
    class_label = "testing"

    # Extract Log-Mel Spectrograms (do not add padding)
    mels = helpers.get_mel_spectrogram(file_path, 0, n_mels=n_mels)
    
    # Save current frame count
    num_frames = mels.shape[1]

    # Add row (feature / label)
    features.append(mels)
    labels.append(class_label)

    # Update frames maximum
    if (num_frames > frames_max):
        frames_max = num_frames

    clear_output(wait=True)
    print("Progress: {}/{}".format(counter+ 1, total_test_samples))
    print("Last file: ", file_path)

    counter += 1

print("Finished: {}/{}".format(counter, total_test_samples))
```
